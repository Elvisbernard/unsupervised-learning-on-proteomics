```python
import pandas as pd
import warnings
from scipy.stats import binom_test, rankdata
import numpy as np
warnings.filterwarnings('ignore')
```

```python
#dfclust=pd.read_csv('clusterv2A.csv')
#dfclust=pd.read_csv('clust_3kA.csv')
#dfclust=pd.read_csv('clusterv_compA.csv')
#dfclust=pd.read_csv('leidenA.csv')
dfclust=pd.read_csv('mergedcluster.csv')
```

```python
dfoutcomeS=pd.read_csv('E:\\Proteomic\\1500_outcome.csv')
dfoutcomeS.rename(columns={'Participant ID':'eid'}, inplace=True)
outcome=dfoutcomeS

#outcome=pd.read_csv('outcomev2.csv')

outcome=pd.read_csv('outcomelong.csv')
```

```python
clustlist=['A'+str(x) for x in range(1,38)]+['B'+str(y) for y in range(1,19)]
```

```python
k=55#int(dfclust.columns[-1].split('cl')[1])
listy=dfclust.columns[1:]
dfratio=pd.DataFrame()
for n in range(1, k+1):
    dfclustt=dfclust#[dfclust['Gender']==0] 
    dfo=dfclustt[['eid', listy[n]]].fillna(0).merge(outcome[['eid']+outcome.columns[2:].tolist()], on='eid')
    dfo.clip(0,1, inplace=True)
    colname=[x for x in dfo.columns[2:] if dfo[x].sum()>14]
    dfo=dfo[dfo.columns[:2].tolist()+colname]
    pos=dfo[dfo[ listy[n]]==1].mean()[2:]
    npos=dfo[dfo[ listy[n]]==1].sum()[2:]
    neg=dfo[dfo[ listy[n]]==0].mean()[2:]
    nn=dfo[dfo[ listy[n]]==0].sum()[2:]
    p=[]
    for i in range(len(pos)):
        p+=[binom_test(x=npos[i], n=dfo[dfo[ listy[n]]==1].shape[0], p=neg[i])]
    rpv=rankdata(p)
    fdr=np.array(p)*len(p)/rpv
    r=pos/neg
    dfratio['prevalence in cluster']=pos
    dfratio['case in cluster']=npos
    dfratio['prevalence out of cluster']=neg
    dfratio['case out of cluster']=nn
    dfratio['odd ratio']=r
    dfratio['raw p-value']=p
    dfratio['p-value Bonferroni']=np.array(p)*len(p)
    dfratio['p-value Benjamini-Hochberg']= fdr #Benjamini-Hochberg
    
    tempdf=dfratio[dfratio['case out of cluster']+dfratio['case in cluster']>14][['prevalence in cluster','case in cluster',
                                                                           'prevalence out of cluster','case out of cluster',
                                                                           'odd ratio', 'raw p-value',
                                                                           'p-value Bonferroni','p-value Benjamini-Hochberg']]
    #tempdf=tempdf[dfratio['case in cluster']>10]
    #tempdf.to_csv('tosend\\'+clustlist[n-1]+'_female_grouped.csv')
    print(clustlist[n-1])
    display(tempdf.dropna(axis=0)[tempdf['raw p-value']<0.05].sort_values(['p-value Bonferroni']).head(20))
    #break
```

```python
k=int(dfclust.columns[-1].split('cl')[1])
dfratio=pd.DataFrame()
for n in range(1, k+1):
    dfclustt=dfclust[dfclust['Gender']==1] #male
    dfo=dfclustt[['eid', 'cl'+str(n)]].fillna(0).merge(outcome[['eid']+outcome.columns[2:].tolist()], on='eid')
    dfo.clip(0,1, inplace=True)
    pos=dfo[dfo['cl'+str(n)]==1].mean()[2:]
    npos=dfo[dfo['cl'+str(n)]==1].sum()[2:]
    neg=dfo[dfo['cl'+str(n)]==0].mean()[2:]
    nn=dfo[dfo['cl'+str(n)]==0].sum()[2:]
    p=[]
    for i in range(len(pos)):
        p+=[binom_test(x=npos[i], n=dfo[dfo['cl'+str(n)]==1].shape[0], p=neg[i])]
    rpv=rankdata(p)
    fdr=np.array(p)*len(p)/rpv
    r=pos/neg
    dfratio['poscl'+str(n)]=pos
    dfratio['nposcl'+str(n)]=npos
    dfratio['negcl'+str(n)]=neg
    dfratio['nnegcl'+str(n)]=nn
    dfratio['ratiocl'+str(n)]=r
    dfratio['pcl'+str(n)]=np.array(p)*len(p)
    dfratio['Bpcl'+str(n)]= fdr #Benjamini-Hochberg
    
    tempdf=dfratio[dfratio['nnegcl'+str(n)]+dfratio['nposcl'+str(n)]>100][['poscl'+str(n),'nposcl'+str(n),
                                                                           'negcl'+str(n),'nnegcl'+str(n),
                                                                           'ratiocl'+str(n),'pcl'+str(n),'Bpcl'+str(n)]]
    tempdf=tempdf[dfratio['nposcl'+str(n)]>10]
    display(tempdf.dropna(axis=0)[tempdf['Bpcl'+str(n)]<0.05].sort_values(['pcl'+str(n)]).head(20))
```
