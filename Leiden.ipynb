```python
def build_df(dfclust,outcome, cl,k):
    dfratio=pd.DataFrame()
    dfo=dfclust[['eid', cl]].fillna(0).merge(outcome[['eid']+outcome.columns[2:].tolist()], on='eid')
    #dfo.clip(0,1, inplace=True)
    pos=dfo[dfo[cl]==k].mean()[2:]
    npos=dfo[dfo[cl]==k].sum()[2:]
    neg=dfo[dfo[cl]!=k].mean()[2:]
    nn=dfo[dfo[cl]!=k].sum()[2:]
    p=[]
    for i in range(len(pos)):
        p+=[binom_test(x=npos[i], n=dfo[dfo[cl]==k].shape[0], p=neg[i])]
    rpv=rankdata(p)
    fdr=np.array(p)*len(p)/rpv
    r=pos/neg
    dfratio['poscl']=pos
    dfratio['nposcl']=npos
    dfratio['negcl']=neg
    dfratio['nnegcl']=nn
    dfratio['ratiocl']=r
    dfratio['pcl']=np.array(p)*len(p) 
    dfratio['Bpcl']= fdr #Benjamini-Hochberg
    
    return  dfratio#[dfratio['nposcl']>3]
 
```

```python
dic={}
for n in ['base']+list(range(9)):
    n=str(n)
    print('clustering: {}'.format(n))
    merged=newdf[['eid',str(n)]].merge(df, on='eid')
    dic[str(n)]={}
    for k in merged[str(n)].unique():
        dfshort=merged[merged[str(n)]==k]
        comp=merged[merged[str(n)]!=k]
    
    
        dic[str(n)][k]={}
        dic[str(n)][k]['high']=[]
        dic[str(n)][k]['low']=[]
        dic[str(n)][k]['count']=dfshort.shape[0]
        #dic[n][k]['n_male']=dfshort.Gender.sum()
        #dic[n][k]['n_female']=dic[cl]['count']-dic[cl]['n_male']

        for var in merged.columns[2:-2]:
            vard=comp[var].describe()

            varc=dfshort[var].describe()
            if varc[4]>vard[6]:
                dic[str(n)][k]['high']+=[var]
            if varc[6]<vard[4]:
                dic[str(n)][k]['low']+=[var]
        dic[str(n)][k]['df']   = build_df(merged, outcome,n, k)
    #dic[n][k]['df_m'] = build_df(dfclust[dfclust.Gender==1], outcome, cl)
    #dic[n][k]['df_f'] = build_df(dfclust[dfclust.Gender==0], outcome, cl)
    
    for k in dic[str(n)].keys():
        print('cluster: '+str(k))
        print(k+': count:'+str(int(dic[n][k]['count']))+'\t: high:'+ str(len(dic[n][k]['high']))+ '\tlow:'+str(len(dic[n][k]['low'])))
        print('intersection:')
        for j in  dic[n].keys():
            h=len(set (dic[n][k]['high']).intersection(set(dic[n][j]['high'])))
            l=len(set (dic[n][k]['low']).intersection(set(dic[n][j]['low'])))
            hl=len(set (dic[n][k]['high']).intersection(set(dic[n][j]['low'])))
            lh=len(set (dic[n][k]['low']).intersection(set(dic[n][j]['high'])))
            try:
                lr=l/len(dic[n][k]['low'])
            except:
                lr=0
            try:
                hr=h/len(dic[n][k]['high'])
            except:
                hr=0

            print(j, 'high:',str(h),'({}%)'.format(round(100*hr)),
                 '\tlow:',str(l),'({}%)'.format(round(100*lr)),
                 '\th vs l:',str(hl),
                 '\tl vs h:',str(lh),)
        print('\n')    
        print('Whole population ({})'.format(dic[n][k]['count']))    
        display(dic[n][k]['df'][dic[n][k]['df']['Bpcl']<0.05].sort_values(['pcl']))    

        print('*-'*60)
```
